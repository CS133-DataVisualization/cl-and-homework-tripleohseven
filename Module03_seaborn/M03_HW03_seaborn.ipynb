In this notebook, we will plot the relationships between different observations of the combat statistics of Pokémon. You will need to import a few packages, including `pandas`, `matplotlib.pyplot`, and `seaborn`. The data is stored in a csv file: https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

Here are the descriptions of this dataset:
- `#` - Pokedex entry number of the Pokemon
- `Name` - name of the Pokemon
- `Type 1` - each Pokemon has a type, this determines weakness/resistance to attacks [referred to as the primary type]
- `Type 2` - some Pokemon are dual type and have 2 [referred to as the secondary type]
- `Total` - sum of all stats that come after this, a general guide to how strong a Pokemon is
- `HP` - hit points, or health, defines how much damage a Pokemon can withstand before fainting
- `Attack` - the base modifier for normal attacks
- `Defense` - the base damage resistance against normal attacks
- `Sp. Atk` - special attack, the base modifier for special attacks
- `Sp. Def` - the base damage resistance against special attacks
- `Speed` - determines which Pokemon attacks first each round
- `Generation` - refers to which grouping/game series the Pokemon was released in
- `Legendary` - a boolean that identifies whether the Pokemon is legendary

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fp = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
```

## Q1: Plot the relationship between Attack and Defense
Create a relationship plot using seaborn's `relplot` function. Plot `Attack` on the x-axis and `Defense` on the y-axis. Use `hue` to display the `HP` (hit point) for each datapoint. You can choose your favorite palette. Include a meaninful title in the plot. After creating the plot, insert a markdown cell :immediately following the plot to describe the relationship between Attack and Defense in this dataset.

### From the plot, I observed....
Results: There is a positive correlation between `Attack` and `Defense`. The hit-points (`HP`) is higher for Pokemons with higher `Attack` and `Defense` values.

```python
# Load the data
df = pd.read_csv(fp)

# Q1: Plot relationship between Attack and Defense with HP as hue
g = sns.relplot(data=df, x='Attack', y='Defense', hue='HP', palette='viridis')
g.fig.suptitle('Relationship Between Attack and Defense (colored by HP)', y=1.02)
plt.show() 
```

## Q2: Plot the relationship between Attack and Total using a hexbin plot
Create a hexbin plot using seaborn's `joinplot` function. Plot `Attack` on the x-axis and `Total` on the y-axis. You can choose the gridsize. Include a title for the plot "Attack vs total for all Pokemon". After generating the plot, insert a markdown cell to answer this question: The hexbin plot displays the relationship between `Attack` and `Total`, what else does this plot show?

```python
# Q2: Hexbin plot of Attack vs Total
g = sns.jointplot(data=df, x='Attack', y='Total', kind='hex', gridsize=20)
g.fig.suptitle('Attack vs Total for all Pokemon', y=1.02)
plt.show()
```

### From the plot, I observed....
The hexbin plot shows a strong positive correlation between Attack and Total stats. Additionally, the hexbin plot displays the **density/concentration** of data points - darker hexagons indicate more Pokemon fall within that range. The marginal histograms on the axes also show the distribution of Attack values (x-axis) and Total values (y-axis) separately, revealing that most Pokemon have Attack values between 50-100 and Total values between 300-500.

## Q3: Create two categorical plots for Pokemon types. `Type 1` and `Type 2` .
For each plot, the unique values of `Type 1` or `Type 2` should be plotted along the y-axis, and the counts of Pokemon should be on the x-axis. Use the color red for the `Type 1` plot and blue for the `Type 2` plot. After the two plots are displayed, insert a markdown cell to write your findings: Which are the most common `Type 1` and `Type 2` Pokemons, and which are the most rare ones?

```python
# Q3: Categorical plots for Type 1 and Type 2
fig, axes = plt.subplots(1, 2, figsize=(14, 8))

# Type 1 plot (red)
sns.countplot(data=df, y='Type 1', color='red', order=df['Type 1'].value_counts().index, ax=axes[0])
axes[0].set_title('Pokemon Count by Type 1')
axes[0].set_xlabel('Count')

# Type 2 plot (blue)
sns.countplot(data=df, y='Type 2', color='blue', order=df['Type 2'].value_counts().index, ax=axes[1])
axes[1].set_title('Pokemon Count by Type 2')
axes[1].set_xlabel('Count')

plt.tight_layout()
plt.show()
```

### Findings on Pokemon Types

**Most Common Types:**
- **Type 1**: Water is the most common primary type, followed by Normal and Grass
- **Type 2**: Flying is the most common secondary type, followed by Ground and Poison

**Most Rare Types:**
- **Type 1**: Flying is the rarest primary type (very few Pokemon have Flying as their primary type)
- **Type 2**: Many Type 2 values are NaN (Pokemon without a secondary type). Among those with Type 2, Ice and Steel appear to be among the rarest

## Q4-6: Your own question
Come up with three more questions to further explore the pokemon dataset.

### My Questions:

**Q4:** How do the overall stats (Total) differ between Legendary and Non-Legendary Pokemon?

**Q5:** What is the distribution of Speed across different Pokemon Generations?

**Q6:** Is there a correlation between Special Attack (Sp. Atk) and Special Defense (Sp. Def)?

```python
# Q4: Legendary vs Non-Legendary Total Stats
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df, x='Legendary', y='Total', palette='Set2', ax=ax)
ax.set_title('Total Stats: Legendary vs Non-Legendary Pokemon')
ax.set_xlabel('Legendary Status')
ax.set_ylabel('Total Stats')
plt.show()
```

```python
# Q5: Speed Distribution Across Generations
fig, ax = plt.subplots(figsize=(12, 6))
sns.violinplot(data=df, x='Generation', y='Speed', palette='muted', ax=ax)
ax.set_title('Speed Distribution Across Pokemon Generations')
ax.set_xlabel('Generation')
ax.set_ylabel('Speed')
plt.show()
```

```python
# Q6: Correlation between Sp. Atk and Sp. Def
g = sns.jointplot(data=df, x='Sp. Atk', y='Sp. Def', kind='reg', color='purple')
g.fig.suptitle('Correlation: Special Attack vs Special Defense', y=1.02)
plt.show()
```

### Findings from Q4-6:

**Q4 Observation:** Legendary Pokemon have significantly higher Total stats compared to non-legendary Pokemon. The median Total for Legendary Pokemon is around 600, while non-legendary Pokemon have a median around 420. This confirms that Legendary Pokemon are indeed more powerful overall.

**Q5 Observation:** Speed distribution is fairly consistent across all generations, with most generations having a similar median speed around 65-70. However, some generations show more variability and outliers, indicating that certain generations introduced both very fast and very slow Pokemon.

**Q6 Observation:** There is a moderate positive correlation between Special Attack and Special Defense. Pokemon with higher Sp. Atk tend to also have higher Sp. Def, though the relationship is not as strong as Attack vs Defense. This suggests that Pokemon specializing in special moves often have balanced special stats.
